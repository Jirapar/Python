```python
## My game is based on Doraemon cartoon Episode 650. Availble online source : https://www.facebook.com/102051841287847/videos/334469744980166/
## DocString for the introduction
""""
    DocString :
    My game is based on a Japanese cartoon named Doraemon. Doraemon was written by Katakana and Hiragana in 1969.
    Doraemon is a famous cartoon in Asia. Doraemon is a male cat that has the magic bag! Doraemon has a best friend named Nobinta. 
    Nobinta is quite an unwise character in the cartoon and he always asks for help from Doraemon. Doraemon can help everything because he has a magic bag!!

    Stage 1: Decide the equipment to catch fish
    Stage 2: How to find the magic bag
    Stage 3: How to catch fish

    Note: There is no bug.

"""
#import random package to use the random.choice for creating replay value
import random

# Create the intro to tell the story of game to understand what happen
def intro():
    
    print("Before starting the game let me know more about you.")
    
    #make user input to have an interaction with the player by asking thier name
    player_name = input(prompt = "What is your Name?: ").capitalize()
    
    #create variable to use as a choices in random.choice
    Adj_1 = ['driving a car', 'walking', 'a boat', 'swimming', 'flying', 'an airplane']
    
    #apply dynamic printing and print the storyline of this game
    #use random.choice for replay value
    print('-'*100)
    print(f"""Hi {player_name}!! Welcome to my game. Doraemon and Nobinta are best friend and they always have an adventure time together. 
For this time Nobinta and Doraemon go to travel with friends on an island by {random.choice(Adj_1)}. After they arrive, they go to catch fish but it has some accidents!!""")
    print('-'*100)

    #apply user input to have an interactive with player
    ready = input(prompt = """Now you are Nobinta, you have mission to solve the problem.
Are you ready to start? <Y/N>:""").lower()

    #Use if-else condition
    if ready == 'yes' in ready or 'y' in ready:
        print("Yeah! Let's go!!!!")
        
        #This is move to stage_1
        stage_1()
    
    else:
        print("Whatever you answer, you must ready now!! Let's go!!")
        
        #This is move to stage_1
        stage_1()

        
# Adding stage_1 and linking it to intro
def stage_1():
    
    print('-'*100)
    print("Let me ask one more question.")
    
    food = input(prompt = "What is the worst food you will never eat it?:") 
    
    #print storyline 
    print('-'*100)
    print(f"""Now you are quite nervous. You are not good at catching fish and you are challenged by other friends.
If you can't catch any fish, you'll get punishment by eating {food}. Oh no!!""")
    print('-'*100)
    
    #print the question number 1
    print("""What equipment do you use to catch fish?
    1.Magical hand fishhook
    2.Worm""")
    
    #apply user input to answer the question number 1
    choice_1 = input(prompt = "Press your answer:\n").lower()
    
    #the condition of each answer for the first question
    if choice_1 == '1'in choice_1 or 'magical' in choice_1 or 'hand' in choice_1 or 'fishhook' in choice_1:

        print("""Oh! That is a cool item from Doraemon! When you use it, the fishhook will find fish from the ocean for you. 
Just wait and relax!! you will get it in 3 seconds.\n """)
        
        
        second = 3
        print(f"Wait for {second} seconds! ")

        #apply while loop to countdown the second
        while True:

            if second > 0:
                second -= 1
                input( prompt ="Press any key three times to countdown for 3 seconds: ")
                print(f"wait for {second} second")
                
            elif second == 0:
                print("You get the strogest fish!!")
                
                #This is move to stage_2
                stage_2()
                break
                
            else:
                print("Something went wrong.")
                #If player not follow the instruction. Go to fail function
                fail()
            
    elif choice_1 == '2' in choice_1 or 'worm' in choice_1:
        print("Oh you may have to wait for 1,000 years to get the fish.")
        
        #If player answer wrong. Go to fail function
        fail()
    
    else:
        print("Something went wrong. You should answer following two choices that I give.")
        
        #If player not follow the instruction. Go to fail function
        fail()

        
# Adding stage_2 and linking it to stage_1      
def stage_2():
    
    #print storyline
    print('-'*100)
    print("""You catch the strongest fish then you can't pull it up!! Doraemon comes to help you.
Then both of you fall into the water and fish pull you into the tube of time-machine. 
Now you go back in time around 500 years ago. Don't worry you have Doraemon who has a magic bag but wait...


The magic bag is lost!!!""")
    #print the question number 2
    print('-'*100)
    print("""How you will find the magic bag?
    1.Diving and find it under the ocean.
    2.Use paddle.""")
    
    #apply user input to answer the question number 2
    choice_2= input(prompt = "Press your answer:\n").lower()
    
    #apply if-else condition for each answer of the question number 2
    if choice_2 == '1' in choice_2 or 'diving' in choice_2 or 'find' in choice_2:
        print("Yes! you find it. It's on the strongest fish that pull you into time-machine. We have to catch it!")
        
        #Right answer. Go to stage_3
        stage_3()
    
    else: 
        print("No way to find it!!")
        #If player answer wrong. Go to fail function
        fail()
        

def stage_3():
    
    #print the storyline
    print('-'*100)
    print("Now you're in the hardest part. If you can't get that magic bag, you can't go home forever!! ")
    print('-'*100)
    
    #print the question number 3
    print("""Which way you choose to catch the strongest fist? Remember you have only one chance.
Hint: This fish like red color
    1. Fighting and catch it by using harpoon.
    2. Group the bamboo and use big rope as the fishing rod and use Doraemon as the bait.""")
    
    #apply user input to answer the question number 3
    choice_3 = input(prompt = "Press your answer:\n").lower()
    
    if choice_3 == '2' in choice_3 or 'group'in choice_3:
        print("""Cool idea! Doraemon has a red tail. You have to persuade Doraemon to be bait. 
Which way do you select? 
        1. Go down on your knee and beg him.
        2. Promise to give one hundred of Dorayaki.""")
        
        #use user input and appply nested conditional statement
        choice_4 = input(prompt = "Press your answer:\n").lower()
        
        if choice_4 == '1' in choice_4 or 'go' in choice_4 or 'knee' in choice_4 or 'beg' in choice_4 :
            print("Doraemon said: 'Definitely no!! I will not be a food of fish'")
            fail()
            
        elif choice_4 == '2' in choice_4 or 'p' in choice_4 or 'dorayaki' in choice_4:
            print ("Doraemon said: 'I will do it! I love Dorayaki!!!'")
            
            #print storyline to explain more detail before the game is over
            print('-'*100)
            print("""Both of you made a fishing rod and use Doraemon as bait. 
When the strongest see the tail of Doraemon, it comes quickly to bite it. 
Then you suddenly pull up Doraemon and the strongest fish! You can catch the fish. Yeah!!""")
            print('-'*100)
            
            #Right answer. Go to last stage to finish game
            last_stage()
            
        else:
            print("Something went wrong.")
            fail()
    else:
        print("This is the strongest fish and it swims very quickly. You can't catch it and you don't get the bag.")
        #If player answer wrong. Go to fail function 
        fail()
                         

def last_stage():
    #print the storyline
    print("""Congrats!! You've done so well. Now Doraemon gets the magic bag back. Let's take a magic door from the magic bag and go to the present time. I am sure that you already miss your home.""")
    input(prompt = "See you! Press any key to end game")
    
def fail(): 
    #print the sentence when you fail from the game
    print('-'*100)
    print("Sorry your mission is failed. Thank you for plaing my game, May be you can try again.")
    
    
    #apply user input in case of you fail and may be want to play game again
    again = input(prompt = "Do you want to play again? <Y/N>\n").lower()
    
    #If yes go back to stage_1 to play again
    if 'yes' in again or 'y' in again:
        stage_1()
    
    #If don't want to play just end the game
    elif 'no' in again or 'n' in again:
        print("Okay, see you! Bye!")
    
    #If something went wrong just end the game    
    else:
        print("Something went wrong")

intro()
```
